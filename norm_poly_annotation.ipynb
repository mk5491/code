```python
import os

def remove_empty_lines(input_file, output_file):
    with open(input_file, 'r') as infile, open(output_file, 'w') as outfile:
        for line in infile:
            if line.strip():  # Check if the line is not empty after stripping whitespace
                outfile.write(line)

if __name__ == "__main__":
    input_directory = 'polygon_annotations'    # Replace with the path to your directory containing .txt files
    output_directory = 'polygon_annotations1'  # Replace with the path where you want to save the output files

    if not os.path.exists(output_directory):
        os.makedirs(output_directory)

    for filename in os.listdir(input_directory):
        if filename.endswith('.txt'):
            input_file_path = os.path.join(input_directory, filename)
            output_file_path = os.path.join(output_directory, filename)

            remove_empty_lines(input_file_path, output_file_path)
```

```python
import os

def convert_polygon_to_yolo_normalized(polygons, img_width, img_height):
    yolo_polygons = []
    for polygon in polygons:
        #print(len(polygon))
        normalized_points = [polygon[0]]
        for i in range(1, len(polygon)+1, 2):
            if i < len(polygon)-1:
                x = polygon[i]
                y = polygon[i + 1]
                normalized_x = x / img_width
                normalized_y = y / img_height
                normalized_points.extend([normalized_x, normalized_y])
        yolo_polygons.append(normalized_points)
    return yolo_polygons

def read_polygons_from_txt(input_file):
    with open(input_file, 'r') as file:
        content = file.read()

    polygons = []
    lines = content.strip().split('\n')
    for line in lines:
        points = [int(coord) for coord in line.split(',') if coord.strip()]
        #print(points)
        polygons.append(points)
#         if len(points) % 2 == 0:  # Check for valid number of coordinates
#             polygons.append(points)
#         else:
#             #print(f"Skipping line: {line}")
#             print(line)

    return polygons

def save_polygons_to_txt(output_file, polygons):
    with open(output_file, 'w') as file:
        for polygon in polygons:
            file.write(f"{polygon[0]:.0f} ")
            for i in range(1, len(polygon), 2):
                if i < len(polygon)-1:
                    file.write(f"{polygon[i]:.6f} {polygon[i + 1]:.6f} ")
            file.write("\n")

if __name__ == "__main__":
    input_directory = 'polygon_annotations1'
    output_directory = 'polygon_norm_annotations1'
    img_width = 1280  # Replace with the actual image width
    img_height = 966  # Replace with the actual image height

    if not os.path.exists(output_directory):
        os.makedirs(output_directory)

    for filename in os.listdir(input_directory):
        if filename.endswith('.txt'):
            input_file_path = os.path.join(input_directory, filename)
            output_file_path = os.path.join(output_directory, filename)

            polygons = read_polygons_from_txt(input_file_path)
            yolo_polygons = convert_polygon_to_yolo_normalized(polygons, img_width, img_height)
            save_polygons_to_txt(output_file_path, yolo_polygons)
```
